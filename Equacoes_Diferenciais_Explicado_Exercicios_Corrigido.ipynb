## Introdução às Equações Diferenciais

As equações diferenciais são equações que envolvem uma função e suas derivadas. Elas são usadas para descrever fenômenos que envolvem taxas de variação, como crescimento populacional, movimento de partículas, variação de temperatura, entre outros. Resolver uma equação diferencial significa encontrar uma função que satisfaça a relação expressa pela equação.

Neste notebook, vamos aprender a resolver equações diferenciais ordinárias (EDOs) usando Python e a biblioteca `scipy.integrate.odeint`. Vamos explicar cada etapa de forma detalhada, para que qualquer pessoa possa entender como funciona a resolução de equações diferenciais numéricas.

## Importando as Bibliotecas Necessárias

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

## Exemplo de Equação Diferencial de Primeira Ordem

Vamos começar resolvendo um exemplo de equação diferencial de primeira ordem:

- \[ \frac{dy}{dt} = -2y \]

Esta é uma equação diferencial linear de primeira ordem. Vamos resolver esta equação usando o método `odeint` da biblioteca `scipy`.

### Importante:
- **Alterar a Equação**: Caso você deseje resolver outra equação, você pode modificar a função `modelo` no código abaixo.

```python
# Definindo o modelo da equação diferencial
def modelo(y, t):
    dydt = -2 * y
    return dydt
```

## Definindo as Condições Iniciais e o Intervalo de Tempo

Para resolver a equação diferencial, precisamos definir:
1. **Condição inicial**: O valor de `y` no tempo `t = 0`.
2. **Intervalo de tempo**: O intervalo no qual queremos resolver a equação.

Vamos definir uma condição inicial de `y(0) = 5` e resolver a equação no intervalo de tempo de `t = 0` a `t = 5`.

```python
# Definindo a condição inicial
y0 = 5

# Definindo o intervalo de tempo
t = np.linspace(0, 5, 100)
```

## Resolvendo a Equação Diferencial

Vamos agora usar a função `odeint` para resolver a equação diferencial. Esta função recebe três parâmetros principais:
1. **modelo**: A função que define a equação diferencial.
2. **y0**: A condição inicial.
3. **t**: O intervalo de tempo para o qual queremos resolver a equação.

Vamos resolver e plotar o resultado.

```python
# Resolvendo a equação diferencial
y = odeint(modelo, y0, t)

# Plotando o resultado
plt.plot(t, y, label='y(t)')
plt.xlabel('Tempo t')
plt.ylabel('y(t)')
plt.title('Solução da Equação Diferencial')
plt.legend()
plt.grid(True)
plt.show()
```

### Explicação dos Parâmetros
- **y0**: A condição inicial da equação diferencial. Altere este valor para ver como diferentes condições iniciais afetam a solução.
- **t**: O intervalo de tempo no qual a equação será resolvida. Altere o valor de `t` para ver o comportamento da solução em diferentes intervalos.

## Exemplo 2: Equação Diferencial de Segunda Ordem

Vamos agora resolver uma equação diferencial de segunda ordem. Para isso, precisamos reescrever a equação de segunda ordem como um sistema de equações de primeira ordem.

Considere a equação diferencial:

- \[ \frac{d^2y}{dt^2} + 3\frac{dy}{dt} + 2y = 0 \]

Podemos reescrever esta equação como um sistema de duas equações de primeira ordem:

1. \[ \frac{dy_1}{dt} = y_2 \]
2. \[ \frac{dy_2}{dt} = -3y_2 - 2y_1 \]

```python
# Definindo o modelo da equação diferencial de segunda ordem
def modelo_segunda_ordem(y, t):
    y1, y2 = y
    dy1dt = y2
    dy2dt = -3 * y2 - 2 * y1
    return [dy1dt, dy2dt]
```

### Definindo as Condições Iniciais e o Intervalo de Tempo

Vamos definir as condições iniciais para `y1(0) = 1` e `y2(0) = 0`, e resolver a equação no intervalo de tempo de `t = 0` a `t = 10`.

```python
# Definindo as condições iniciais
y0_segunda_ordem = [1, 0]

# Definindo o intervalo de tempo
t_segunda_ordem = np.linspace(0, 10, 100)
```

## Resolvendo a Equação Diferencial de Segunda Ordem

Vamos agora usar a função `odeint` para resolver o sistema de equações de primeira ordem.

```python
# Resolvendo a equação diferencial de segunda ordem
solucao_segunda_ordem = odeint(modelo_segunda_ordem, y0_segunda_ordem, t_segunda_ordem)

# Plotando o resultado
plt.plot(t_segunda_ordem, solucao_segunda_ordem[:, 0], label='y1(t)')
plt.xlabel('Tempo t')
plt.ylabel('y1(t)')
plt.title('Solução da Equação Diferencial de Segunda Ordem')
plt.legend()
plt.grid(True)
plt.show()
```

## Exercícios Adicionais

### Exercício 1: Equação Diferencial de Primeira Ordem
Considere a equação diferencial de primeira ordem:
- \[ \frac{dy}{dt} = 3y - 5 \]
Resolva esta equação diferencial para `y(0) = 2` no intervalo de tempo de `t = 0` a `t = 5`.

```python
# Definindo o modelo da equação diferencial do exercício 1
def modelo_ex1(y, t):
    dydt = 3 * y - 5
    return dydt

# Definindo a condição inicial
y0_ex1 = 2

# Definindo o intervalo de tempo
t_ex1 = np.linspace(0, 5, 100)

# Resolvendo a equação diferencial
y_ex1 = odeint(modelo_ex1, y0_ex1, t_ex1)

# Plotando o resultado
plt.plot(t_ex1, y_ex1, label='y(t)')
plt.xlabel('Tempo t')
plt.ylabel('y(t)')
plt.title('Solução da Equação Diferencial do Exercício 1')
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 2: Equação Diferencial de Segunda Ordem
Considere a equação diferencial de segunda ordem:
- \[ \frac{d^2y}{dt^2} + 4\frac{dy}{dt} + 4y = 0 \]
Resolva esta equação diferencial para `y1(0) = 2` e `y2(0) = -1` no intervalo de tempo de `t = 0` a `t = 10`.

```python
# Definindo o modelo da equação diferencial do exercício 2
def modelo_ex2(y, t):
    y1, y2 = y
    dy1dt = y2
    dy2dt = -4 * y2 - 4 * y1
    return [dy1dt, dy2dt]

# Definindo as condições iniciais
y0_ex2 = [2, -1]

# Definindo o intervalo de tempo
t_ex2 = np.linspace(0, 10, 100)

# Resolvendo a equação diferencial
solucao_ex2 = odeint(modelo_ex2, y0_ex2, t_ex2)

# Plotando o resultado
plt.plot(t_ex2, solucao_ex2[:, 0], label='y1(t)')
plt.xlabel('Tempo t')
plt.ylabel('y1(t)')
plt.title('Solução da Equação Diferencial do Exercício 2')
plt.legend()
plt.grid(True)
plt.show()
```
